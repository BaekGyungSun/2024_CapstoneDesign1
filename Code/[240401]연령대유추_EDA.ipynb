```python
# 남자
import librosa
import librosa.display
import matplotlib.pyplot as plt

# 음성 파일 로드
file_path = 'C:/Users/wnsdn/Desktop/SCH/2024-1/Practice/CapstoneDesign/자유대화 음성(일반남녀)/Training/[원천]1.AI챗봇_1/일반남여_일반통합05_M_1455308554-6_37_수도권_실내/일반남여_일반통합05_M_1455308554-6_37_수도권_실내_06313.wav'  # 분석하고자 하는 파일의 경로
audio_data_m, sample_rate = librosa.load(file_path)

# 기본 정보 출력
print(f"Sample rate: {sample_rate} Hz")
print(f"Audio duration: {len(audio_data_m) / sample_rate:.2f} seconds")

# 파형 시각화
plt.figure(figsize=(14, 5))
librosa.display.waveshow(audio_data_m, sr=sample_rate)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# 남자
# 스펙트로그램 시각화
X = librosa.stft(audio_data_m)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')
plt.colorbar()
plt.title('Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

```python
# 오디오 길이와 특성 간의 관계

import seaborn as sns

sns.pairplot(audio[['Duration (Seconds)', 'Average Volume', 'Zero Crossing Rate', 'Spectral Centroid']], diag_kind='kde')
plt.show()
```

```python
# 나이 그룹별 평균 볼륨 분석
audio.groupby('Age Group')['Average Volume'].mean().plot(kind='bar')
plt.title('Average Volume by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Volume')
plt.show()
```

```python
# 다변량 분석

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = audio['Sample Rate (Hz)']
y = audio['Duration (Seconds)']
z = audio['Average Volume']

ax.scatter(x, y, z, c='r', marker='o')
ax.set_title('3D Plot of Sample Rate, Duration, and Average Volume')
ax.set_xlabel('Sample Rate (Hz)')
ax.set_ylabel('Duration (Seconds)')
ax.set_zlabel('Average Volume')

plt.show()
```

```python
# 오디오 특성과 나이 간의 상관 관계

correlation = audio[['Duration (Seconds)', 'Average Volume', 'Zero Crossing Rate', 'Spectral Centroid', 'Spectral Bandwidth', 'Spectral Roll-off', 'First MFCC']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Audio Features')
plt.show()
```

```python
# 오디오 길이 및 특성에 따른 클러스터링

from sklearn.cluster import KMeans

# 클러스터링을 위한 특성 선택
features_for_clustering = audio[['Duration (Seconds)', 'Average Volume', 'Spectral Centroid']]

# KMeans 클러스터링
kmeans = KMeans(n_clusters=3)
audio['Cluster'] = kmeans.fit_predict(features_for_clustering)

# 클러스터 결과 시각화
sns.scatterplot(x='Duration (Seconds)', y='Average Volume', hue='Cluster', data=audio)
plt.title('Clustering of Audio Files by Duration and Average Volume')
plt.xlabel('Duration (Seconds)')
plt.ylabel('Average Volume')
plt.show()
```

```python
# 성별에 따른 MFCCs 분석
sns.boxplot(x='Gender', y='First MFCC', data=audio)
plt.title('First MFCC Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('First MFCC')
plt.show()
```

```python
# 연령대별 제로 크로싱 레이트 분석
sns.boxplot(x='Age Group', y='Zero Crossing Rate', data=audio)
plt.title('Zero Crossing Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Zero Crossing Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# 스펙트럴 센트로이드와 밴드위드의 관계
sns.scatterplot(x='Spectral Centroid', y='Spectral Bandwidth', hue='Gender', data=audio)
plt.title('Spectral Centroid vs Spectral Bandwidth by Gender')
plt.xlabel('Spectral Centroid')
plt.ylabel('Spectral Bandwidth')
plt.show()
```

```python
# 성별과 연령대별 오디오 길이의 다중 변수 분석
sns.catplot(x='Age Group', y='Duration (Seconds)', hue='Gender', kind='bar', data=audio, ci=None)
plt.title('Audio Length by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Duration (Seconds)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 클러스터링 결과와 성별 비교
sns.countplot(x='Cluster', hue='Gender', data=audio)
plt.title('Cluster Distribution by Gender')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()

# 클러스터링 결과와 연령대 비교
sns.countplot(x='Cluster', hue='Age Group', data=audio)
plt.title('Cluster Distribution by Age Group')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```
